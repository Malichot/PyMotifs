## Transformation en motifs : 

Le présent script prend en entrée le script d'annotation. Il permet de transformer les textes annotés en motifs.

```python
import pandas as pd
import numpy as np
```

```python
# Reading csv : 

df = pd.read_csv("/Users/adesacy/Desktop/PyMotifs/output/corpus_annotated_test.csv", index_col=0)
df.head(50)
```

## Quelques nettoyages préalables : 

```python
df['lemma'] = df['lemma'].str.replace("\\['", '', regex = True)
df['lemma'] = df['lemma'].str.replace("'\\]", '', regex = True)

df['pos'] = df['pos'].str.replace("\\['", '', regex = True)
df['pos'] = df['pos'].str.replace("'\\]", '', regex = True)

df['morph'] = df['morph'].str.replace("\\[", '', regex = True)
df['morph'] = df['morph'].str.replace("\\]", '', regex = True)
df.head(50)
```

```python
# Removing double tags :

df.loc[df.words == "d'", ["lemma", "pos"]] = 'de', 'DET'
df.loc[df.words == "l'", ["lemma", "pos"]] = 'le', 'DET'
df.loc[df.words == "qu'", ["lemma", "pos"]] = 'que', 'SCONJ'
df.loc[df.words == "s’", ["lemma", "pos"]] = 'se', 'PRON'

df.head(50)
```

## Mots invariables, locutions adverbiales, verbes courants, titres :

On souhaite conserver certains de ces termes. Il faut ici se fonder sur les lemmes (c'est la colonne que nous garderons à la fin et qui sera celle des motifs) puis transformer les pos-tag pour qu'ils ne soient pas changés ensuite au cours des transformations futures. Ex : on veut garder le verbe être pour qu'il demeure dans les motifs. Pour cela, obligation de transformer le POS.

```python
mots_inv = pd.read_csv('/Users/adesacy/Desktop/PyMotifs/data/mots_invariables.txt',
                       sep=',',header=None)[0].tolist()

locutions = pd.read_csv('/Users/adesacy/Desktop/PyMotifs/data/locutions_adverbiales.txt',
                       sep = ",", header=None)[0].tolist()

verbes_courants = pd.read_csv('/Users/adesacy/Desktop/PyMotifs/data/verbes_courants.txt',
                       sep = ",", header=None)[0].tolist()

titres = pd.read_csv('/Users/adesacy/Desktop/PyMotifs/data/titres.txt',
                       sep = ",", header=None)[0].tolist()
```

```python
# On remplace les POS par les mots pour conserver ces mots invariables dans les 
# conservations futures des POS pour les motifs : 

idminv = df["words"].isin(mots_inv)

# Check wich is True ie which POS are to be replaced by their lemma value

pos_to_replace = np.where(idminv)[0]

# Replace them : 

df['pos'][pos_to_replace] = df['words'][pos_to_replace]



# Idem avec les locutions :

id_locutions = df['lemma'].isin(locutions)
pos_to_replace = np.where(id_locutions)[0]
df['pos'][pos_to_replace] = df['lemma'][pos_to_replace]

# Idem avec les verbes courants (lemma) : 

id_vb_courants = df['lemma'].isin(verbes_courants)
pos_to_replace = np.where(id_vb_courants)[0]
df['pos'][pos_to_replace] = df['lemma'][pos_to_replace]

# Idem avec les titres (lemma) : 

id_titres = df['lemma'].isin(titres)
pos_to_replace = np.where(id_titres)[0]
df['pos'][pos_to_replace] = df['lemma'][pos_to_replace]

print(df.head(50))
```
